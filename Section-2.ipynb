```python
import seaborn as sns
import pandas as pd
```

```python
# def readPDB
# Team Member: Joseph Rocha
# Purpose: reads a pdb file into a list
# Parameters:
#            path - The path of the file to be read
# Returns: a list of lines
# Notes: The path parameter will be replaced with a link.
#        Open will be replaced by a URL request.
```

```python
def readPDB(path):
    file = open(path, "r")
    lines = file.readlines()
    lines = [line.strip("\n") for line in lines]
    return lines
```

```python
# def returnSection
# Team Member: Joseph Rocha
# Purpose: returns a list of lines that start with the given
# Parameters:
#            lines - lines to be
#            start - the string that the line should start with
# Returns: a list of lines that start with start.
```

```python
def returnSection(lines, start):
    return [line for line in lines if line.startswith(start)]
```

```python
#WARNING! Amino Acid 202 is missing in PDB, but we don't handle it here, should we skip???
def getSEQRESDf(seqres):
    df = pd.DataFrame(columns=['serNum', 'chainID', 'numRes', 'seqNum', 'residue'])
    for line in seqres:
        sequence = line[19:].split()
        for idx, residue in enumerate(sequence):
            serNum = line[7:11].strip() # COLUMNS 8 - 10, serNum
            chainID = line[11].strip() #COLUMNS 11, chainID
            numRes = line[13:17].strip() #COLUMNS, numRes
            seqNum = (idx+1)+((int(serNum)-1)*13)
            df = df.append({'serNum': serNum,
                            'chainID': chainID,
                            'numRes': numRes,
                            'seqNum': seqNum,
                            'residue': residue}, ignore_index=True)
    return df
```

```python
def getBetaSheetDf(strands):
    df = pd.DataFrame(columns=['strand', 
                               'numStrands',
                               'initChainID',
                               'initSeqNum',
                               'endChainID',
                               'endSeqNum'])
    for line in strands:
        strand = line[7:10].strip()
        numStrands = line[14:16].strip()
        initChainID = line[21].strip()
        initSeqNum = line[22:26].strip()
        endChainID = line[32].strip()
        endSeqNum = line[33:37].strip()
        df = df.append({'strand': strand,
                        'numStrands': numStrands,
                        'initChainID': initChainID,
                        'initSeqNum': initSeqNum,
                        'endChainID': endChainID,
                        'endSeqNum': endSeqNum}, ignore_index=True)
    return df
```

```python
def getHelixDf(strands):
    df = pd.DataFrame(columns=['serNum', 
                               'initChainID',
                               'initSeqNum',
                               'endChainID',
                               'endSeqNum',
                               'helixClass',
                               'length'])
    for line in strands:
        serNum = line[7:10].strip()
        initChainID = line[19].strip()
        initSeqNum = line[21:25].strip()
        endChainID = line[31].strip()
        endSeqNum = line[33:37].strip()
        helixClass = line[38:40].strip()
        length = line[71:76].strip()
        df = df.append({'serNum': serNum,
                        'initChainID': initChainID,
                        'initSeqNum': initSeqNum,
                        'endChainID': endChainID,
                        'endSeqNum': endSeqNum,
                        'helixClass': helixClass,
                        'length': length}, ignore_index=True)
    return df
```

```python
lines = readPDB("3uts.pdb")
helix = returnSection(lines, "HELIX")
helix_df = getHelixDf(helix)
helix_df.head()
```

```python
sheets = returnSection(lines, "SHEET")
sheets_df = getBetaSheetDf(sheets)
sheets_df.head()
```

```python
seqres = returnSection(lines, "SEQRES")
seqres_df = getSEQRESDf(seqres)
seqres_df.head()
```
